```python
x=df.iloc[:,:-1].values  #input_data
y=df.iloc[:,-1].values   #output_data
y
```

```python
#training data and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
y_test
```

```python
#normalization==>ellam ore rangillott mattan ayitt
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=5)
classifier.fit(x_train,y_train)
y_predicit=classifier.predict(x_test)
#y_predicit
print(classifier.predict([[2,197,70,45,543,30.5,0.158,53]]))
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
result=confusion_matrix(y_test,y_predicit)
# result
score=accuracy_score(y_test,y_predicit)
score
```
